```python
# Performance Visualization
# Comparing Traditional ETL vs DMM

# Imports
import matplotlib.pyplot as plt
import pandas as pd

# Define benchmark data
data = pd.DataFrame({
    'Method': ['Traditional ETL', 'DMM'],
    'Execution Time (sec)': [0.8799, 0.8117],
    'Memory Usage (MB)': [37.14, 0.06]
})
data
```

```python
# Plot execution time
plt.figure(figsize=(6, 4))
plt.bar(data['Method'], data['Execution Time (sec)'], color=['skyblue', 'lightgreen'])
plt.title('Execution Time Comparison')
plt.ylabel('Time (seconds)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Plot memory usage
plt.figure(figsize=(6, 4))
plt.bar(data['Method'], data['Memory Usage (MB)'], color=['salmon', 'blue'])
plt.title('Memory Usage Comparison')
plt.ylabel('Memory (MB)')
plt.grid(axis='y')
plt.tight_layout()
k = [0.06, 5, 10 ,15, 20, 25, 30, 35, 37.14, 40]
plt.yticks(k)
plt.show()
```
