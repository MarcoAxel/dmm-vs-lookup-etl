##  ETL Performance Comparison
### Comparing Traditional ETL, Regular DMM, and Compacted DMM

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'Method': ['Traditional ETL', 'Regular DMM', 'Compacted DMM'],
    'Execution Time (sec)': [0.5780, 0.5468, 0.4611],
    'Memory Usage (MB)': [0.20, 0.03, 0.02]
})
data
```

```python
plt.figure(figsize=(6, 4))
plt.bar(data['Method'], data['Execution Time (sec)'], color=['skyblue', 'gold', 'lightgreen'])
plt.title('Execution Time Comparison')
plt.ylabel('Time (seconds)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.bar(data['Method'], data['Memory Usage (MB)'], color=['salmon', 'orange', 'mediumseagreen'])
plt.title('Memory Usage Comparison')
plt.ylabel('Memory (MB)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['Traditional ETL', 'Regular DMM', 'Compacted DMM']
datasetT = [0.5780, 0.5468, 0.4611]   # Time
datasetM = [0.20, 0.03, 0.02]         # Memory

x = np.arange(len(labels))  # positions for the bars
bar_width = 0.35

fig, ax1 = plt.subplots(figsize=(9,6))

# --- Bars for Time (left y-axis) ---
bars1 = ax1.bar(x - bar_width/2, datasetT, width=bar_width,
                color=["#4CDAF3", '#FFD580', 'lightgreen'])
ax1.set_ylabel('Time (sec)', fontsize=14, color="blue")
ax1.tick_params(axis='y', labelcolor="blue")

# --- Second y-axis for Memory ---
ax2 = ax1.twinx()
bars2 = ax2.bar(x + bar_width/2, datasetM, width=bar_width,
                color=["#0F41E5", "#E7A218", 'green'])
ax2.set_ylabel('Memory (GB)', fontsize=14, color="darkgreen")
ax2.tick_params(axis='y', labelcolor="darkgreen")

# --- Force both y-axes to share the same scale ---
ymin = 0
ymax = max(max(datasetT), max(datasetM)) * 1.1
ax1.set_ylim(ymin, ymax)
ax2.set_ylim(ymin, ymax)

# --- Custom x-axis labels with two lines ---
custom_labels = [f"{lbl}\n(Left: Time, Right: Memory)" for lbl in labels]
ax1.set_xticks(x)
ax1.set_xticklabels(custom_labels, fontsize=13)

# Title
ax1.set_title('Time vs Memory Usage in ETL Approaches', fontsize=20)

plt.tight_layout()
plt.show()
```
