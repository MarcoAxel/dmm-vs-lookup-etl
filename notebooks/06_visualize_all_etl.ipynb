##  ETL Performance Comparison
### Comparing Traditional ETL, Regular DMM, and Compacted DMM

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'Method': ['Traditional ETL', 'Regular DMM', 'Compacted DMM'],
    'Execution Time (sec)': [0.5780, 0.5468, 0.4611],
    'Memory Usage (MB)': [0.20, 0.03, 0.02]
})
data
```

```python
plt.figure(figsize=(6, 4))
plt.bar(data['Method'], data['Execution Time (sec)'], color=['skyblue', 'gold', 'lightgreen'])
plt.title('Execution Time Comparison')
plt.ylabel('Time (seconds)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.bar(data['Method'], data['Memory Usage (MB)'], color=['salmon', 'orange', 'mediumseagreen'])
plt.title('Memory Usage Comparison')
plt.ylabel('Memory (MB)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## Comparing the two DMM implementations

```python
dmm_data = pd.DataFrame({
    'Method': ['Regular DMM', 'Compacted DMM'],
    'Execution Time (sec)': [0.5468, 0.4611],
    'Memory Usage (MB)': [0.03, 0.02]
})
dmm_data
```

```python
plt.figure(figsize=(6, 4))
plt.bar(dmm_data['Method'], dmm_data['Execution Time (sec)'], color=['gold', 'lightgreen'])
plt.title('DMM Execution Time Comparison')
plt.ylabel('Time (seconds)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.bar(dmm_data['Method'], dmm_data['Memory Usage (MB)'], color=['orange', 'mediumseagreen'])
plt.title('DMM Memory Usage Comparison')
plt.ylabel('Memory (MB)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
